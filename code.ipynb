```python
#using linear regression model

import pandas as pd
from sklearn import linear_model
import numpy as np  
import warnings 
import matplotlib.pyplot as plt

df = pd.read_csv("data.csv")

X = df[['pH','Temperature','Taste','Odour','Fat' ,'Turbidity','Colour']]
y = df['Grade']

regr = linear_model.LinearRegression()
regr.fit(X, y)

input_data = pd.DataFrame([[6.1, 50, 0, 1, 1, 0, 243]], 
                          columns=['pH', 'Temperature', 'Taste', 'Odour', 'Fat', 'Turbidity', 'Colour'])
predicted = regr.predict(input_data)

print("your output :" , predicted)

print("\n Coefficients::", regr.coef_)

import matplotlib.pyplot as plt

# Predict on the entire dataset
predicted = regr.predict(X)

print("\n all predicted values::",predicted)

# Create a scatter plot of actual vs predicted values
plt.scatter(y, predicted, color='blue')
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.title('Actual vs Predicted Values')

# Plot a line representing perfect predictions
plt.plot([y.min(), y.max()], [y.min(), y.max()], color='red', linewidth=2)

plt.show()
```
